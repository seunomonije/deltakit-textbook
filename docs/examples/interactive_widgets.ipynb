# Interactive widgets

Example of interactive widgets with Matplotlib

```python
from __future__ import annotations

import holoviews as hv
import numpy as np
from holoviews import Store
from IPython import get_ipython

get_ipython().enable_matplotlib(gui="inline")

hv.notebook_extension("matplotlib")
hv.output(widget_location='bottom')

Store.renderers["matplotlib"].fig = "svg"
Store.renderers["matplotlib"].dpi = 100
```

```python
def make_plot(alpha):
    return hv.Path(([1,2,3], np.array([1,2,3]) * alpha))

hv.HoloMap(
    {
        alpha: make_plot(alpha) for alpha in np.arange(1, 10., 0.5)
    },
    kdims=["alpha"],
)
```
